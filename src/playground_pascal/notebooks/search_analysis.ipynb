```python
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
from scipy.stats import linregress

init_notebook_mode(True)


def get_df(fname):
    return pd.read_csv("../summaries/" + fname)


def plot_regression_df(df):
    x = df.index
    acc = df['mean_last_epoch_val_acc']
    acc_scatter = go.Scatter(x=x, y=acc, mode='markers', name='acc observartions')

    regr = linregress(x, acc)
    y_pred = regr.slope * x + regr.intercept
    print(regr.slope)
    print(regr.intercept)
    acc_pred_scatter = go.Scatter(x=x, y=y_pred, mode='lines', name='acc regression')
    return iplot([acc_scatter, acc_pred_scatter])


def plot_regression(fname):
    df = get_df(fname)
    plot_regression_df(df)
```

```python
df = get_df('search_summary_bayesian_1.csv')
x = df.index.tolist()
acc = df['mean_last_epoch_val_acc']
acc_scatter = go.Scatter(x=x, y=acc, mode='markers', name='acc observartions')
regr = linregress(x, acc)
#y_pred = regr.slope * x + regr.intercept
regr.intercept
```
