```python
import matplotlib.pyplot as plt
import numpy as np
from bayes_opt import BayesianOptimization
from matplotlib import cm
```

```python
def target(x, y):
    return (np.exp(-(x - 2) ** 2) + np.exp(-(x - 6) ** 2 / 10) + 1 / (x ** 2 + 1)) + 1 / 100 * ((y - 3) ** 2 + 1)
```

```python

fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-2, 10, 0.01)
Y = np.arange(-2, 10, 0.01)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
Z = target(X, Y)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

```python
bo = BayesianOptimization(target, {'x': (-2, 10), 'y': (-2, 10)})
bo.maximize(init_points=2, n_iter=0, acq='ucb', kappa=5)
```
